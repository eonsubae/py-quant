```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
```

```python
# x축은 -10 ~ 10 사이이며 간격은 0.001
```

```python
x_axis = np.arange(-10, 10, 0.001)
```

```python
# 평균 = 0, 표준편차 = 2.0인 정규분포를 생성
```

```python
plt.plot(x_axis, norm.pdf(x_axis, 0, 2))
```
